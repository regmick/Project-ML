## Churn Modeling using Artificial Neural Network (ANN)

Here I am using the fake dataset but the features in the dataset represents the real features
that are in banking data. The objective of this project is building a ANN model to predict the
customer churn in business. This is the classification problem. I am going to use keras library to build the ANN model here.

```python
# First import some libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Always the first part of model building is data preprocessing. Lets do this.
# Importing data and choosing the features. Since first 3 columns in the dataset has 
# no effect of decision so I am selecting features from 4th column and target column is 
# last one.
data_set = pd.read_csv('Churn_Modelling.csv')
X = data_set.iloc[:, 3:13].values
y = data_set.iloc[:, 13].values
```

```python
data_set.head()
```

```python
# Now dataset shows that 'Geography' and 'Gender' as a categorical variables.
# Lets encode them. First encode the 'Geography' column(index 1)
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_1 = LabelEncoder()
X[:, 1] = labelencoder_1.fit_transform(X[:, 1])
```

```python
# Similarly change the 'Gender' column (index 2)
labelencoder_2 = LabelEncoder()
X[:, 2] = labelencoder_2.fit_transform(X[:, 2])
```

```python
# Now lets check X
X
```

```python
# Create the dummy variables for 'Geography" and select all columns
# in X except first colume to remove one dummy variable.
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:, 1:]
```

```python
X
```

```python
# Splitting the dataset into the Training set and Test set (80 % and 20 %)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 101)
```

```python
# Now next important part in data preprocessing is feature scaling.
# Feature Scaling require in neural net models
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# I finished by preprocessing the data set which I am going to use in my ANN model later
# Now next part is building ANN model using keras library. Lets build ANN model.
```

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initialising the ANN classifier
classifier = Sequential()
```

```python
# Adding the input layer (11 nodes) and the first hidden layer (6 nodes here)
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
```

```python
# Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
```

```python
# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
# Predicting the Test set results. The output is probability because our output function
# is sigmoid. Now we need to choose threshold and say that above 0.5 is True and below is 
# False (ie compare as 1 and 0)
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

## Model Evaluation: k-fold cross validation

```python
# Model evaluation using keras and sklearn
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

```python
# Lets define a function and use above used commands to build ANN model inside a function
def build_classifier(optimizer):
    classifier = Sequential()
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
```

```python
# Here I am going to build a classifier but through above function.
classifier = KerasClassifier(build_fn = build_classifier)
```

```python
# I am going to use k-fold cross validation here. Also I am doing parameter tunning
# so lets define parameters for that
parameters = {'batch_size': [20, 32],
              'epochs': [100, 150],
              'optimizer': ['adam', 'rmsprop']}
grid_search = GridSearchCV(estimator = classifier,
                           param_grid = parameters,
                           scoring = 'accuracy',
                           cv = 10)
grid_search = grid_search.fit(X_train, y_train)
best_parameters = grid_search.best_params_
best_accuracy = grid_search.best_score_
```
